```python
def return_correlated_features(df, lower_threshold, upper_threshold):
    corr_matrix = df.corr()

    #the matrix is symmetric so we need to extract upper triangle matrix without diagonal (k = 1)

    sol = (corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
                    .stack()
                    .sort_values(ascending=False))

    filtered_sol = sol[(sol >= lower_threshold) & (sol <= upper_threshold)]

    filtered_sol_df = pd.DataFrame(filtered_sol)
    filtered_sol_df.rename({0: "Correlation"}, inplace=True)

    return filtered_sol_df
```

```python
def plot_3_plots(df, x1, y1, x2, y2, x3, y3, x4, y4, x5, y5, x6, y6):

    corr_matrix = df.corr()

    # Create a figure with 3 subplots in one row
    fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(14, 8))

    # Plot data on each subplot and set unique labels and titles
    axs[0, 0].scatter(x1, y1, data=df)
    axs[0, 0].set_xlabel('{}'.format(x1))
    axs[0, 0].set_ylabel('{}'.format(y1))
    axs[0, 0].text(10, 14, 'Correlation= {}'.format(round(corr_matrix[x1][y1], 2)), fontsize=8, color='black', ha='left', va='top', bbox=dict(facecolor='white', edgecolor='none', pad=1))

    # axs[0, 0].set_title("{} by {}".format(y1, x1))

    axs[0, 1].scatter(x2, y2, data=df)
    axs[0, 1].set_xlabel('{}'.format(x2))
    axs[0, 1].set_ylabel('{}'.format(y2))
    axs[0, 1].text(1, 9, 'Correlation= {}'.format(round(corr_matrix[x2][y2], 2)), fontsize=8, color='black', ha='left', va='top', bbox=dict(facecolor='white', edgecolor='none', pad=1))

    # axs[0, 1].set_title("{} by {}".format(y2, x2))

    axs[0, 2].scatter(x3, y3, data=df)
    axs[0, 2].set_xlabel('{}'.format(x3))
    axs[0, 2].set_ylabel('{}'.format(y3))
    axs[0, 2].text(4, 4, 'Correlation= {}'.format(round(corr_matrix[x3][y3], 2)), fontsize=8, color='black', ha='right', va='top', bbox=dict(facecolor='white', edgecolor='none', pad=1))
    # axs[0, 2].set_title("{} by {}".format(y3, x3))

    axs[1, 0].scatter(x4, y4, data=df)
    axs[1, 0].set_xlabel('{}'.format(x4))
    axs[1, 0].set_ylabel('{}'.format(y4))
    axs[1, 0].text(2, 0, 'Correlation= {}'.format(round(corr_matrix[x4][y4], 2)), fontsize=8, color='black', ha='right', va='top', bbox=dict(facecolor='white', edgecolor='none', pad=1))
    # axs[1, 0].set_title("{} by {}".format(y4, x4))

    axs[1, 1].scatter(x5, y5, data=df)
    axs[1, 1].set_xlabel('{}'.format(x5))
    axs[1, 1].set_ylabel('{}'.format(y5))
    axs[1, 1].text(1, 20, 'Correlation= {}'.format(round(corr_matrix[x5][y5], 2)), fontsize=8, color='black', ha='right', va='top', bbox=dict(facecolor='white', edgecolor='none', pad=1))
    # axs[1, 1].set_title("{} by {}".format(y5, x5))

    axs[1, 2].scatter(x6, y6, data=df)
    axs[1, 2].set_xlabel('{}'.format(x6))
    axs[1, 2].set_ylabel('{}'.format(y6))
    axs[1, 2].text(0.4, 7, 'Correlation= {}'.format(round(corr_matrix[x6][y6], 2)), fontsize=8, color='black', ha='right', va='top', bbox=dict(facecolor='white', edgecolor='none', pad=1))
    # axs[1, 2].set_title("{} by {}".format(y6, x6))

    #fig.tight_layout()
    # Show the figure
    plt.savefig("../../results/images/Correlation_Figures.jpg")
    plt.show()
```
